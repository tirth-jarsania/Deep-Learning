Plotting Sigmoid


# 2-D plotting

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clr
```

```
def sigmoid( x,w,b ):
  return 1 / ( 1 + np.exp(-(x*w + b)))
  
```

```
w = -1   #@param {type: "slider", min: -10, max: 10}
b = 2     #@param {type: "slider", min: -5, max: 5}
X = np.linspace(-10,10,100)
Y = sigmoid(X,w,b)

plt.plot(X,Y)
plt.show()
```

3d plotting to visulize sigmoid function on two features

```
from mpl_toolkits import mplot3d
```

```
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)

X1, X2 = np.meshgrid(x1, x2)
w1 = 0.2
w2 = 0.8
b = 2
def sigmoid2D( x1,x2,w1,w2,b ):
  return 1 / ( 1 + np.exp(-(x1*w1 +x2*w2+ b)))

Y = sigmoid2D(X1,X2,w1,w2,b)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, Y, 200, cmap='cubehelix')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

```

```
my_cmap = clr.LinearSegmentedColormap.from_list("",["red","violet" , "blue"])
plt.contourf(X1,X2,Y,cmap = my_cmap , alpha = 0.6)
plt.show()
```

Compute loss for given dataset

```
X = np.random.random(25)*20 - 10
print( X )
```

```
w_unknown = 0.5
b_unknown = 0.25
print(X)
Y = sigmoid(X,w_unknown , b_unknown)
plt.plot( X,Y,'*' )
plt.show()
```

```
def loss_func( X,Y,w_est,b_est ):
  loss = 0
  for x,y in zip(X,Y):
    loss += (y - sigmoid(x,w_est,b_est))**2
  return loss
```

```
W = np.linspace(-1,1,100)
B = np.linspace(-1,1,100)

WW,BB = np.meshgrid(W,B)
Loss = np.zeros(WW.shape)
```

```
WW.shape
```

```
for i in range(WW.shape[0]):
  for j in range(WW.shape[1]):
    Loss[i,j] = loss_func(X,Y,WW[i,j],BB[i,j])
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(WW, BB, Loss, 200, cmap='viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('loss')

#ax.view_init(30,0)
```

```
cr = np.argmin(Loss)
row_index = cr//(Loss.shape[1])
column_index = cr % (Loss.shape[1])

print(row_index , column_index)
```
